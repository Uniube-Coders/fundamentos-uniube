### M√≠nimo M√∫ltiplo Comum ($MMC$) e M√°ximo Divisor Comum ($MDC$)

#### 1. **M√≠nimo M√∫ltiplo Comum (MMC)**

[Clique aqui para ver o v√≠deo üé¨üé•üî¥‚ñ∂](https://www.youtube.com/)


O M√≠nimo M√∫ltiplo Comum de dois ou mais n√∫meros √© o menor n√∫mero inteiro positivo que √© m√∫ltiplo de todos esses n√∫meros. O MMC √© √∫til para resolver problemas que envolvem sincroniza√ß√£o de eventos, alinhamento de ciclos ou combina√ß√£o de diferentes fra√ß√µes.

##### **Defini√ß√£o Formal:**
Seja $a$ e $b$ dois inteiros positivos. O M√≠nimo M√∫ltiplo Comum de $a$ e $b$, denotado por $\text{MMC}(a, b)$, √© o menor n√∫mero inteiro positivo $m$ tal que:

$$
m \div a = k \quad \text{e} \quad m \div b = l
$$

onde $k$ e $l$ s√£o inteiros positivos.

##### **Como Encontrar o MMC:**
1. **Decomposi√ß√£o em Fatores Primos:**
   - Primeiro, decompomos cada n√∫mero em seus fatores primos.
   - Exemplo: $12 = 2^2 \times 3$ e $15 = 3 \times 5$.

2. **Escolha dos Maiores Expoentes:**
   - Para cada fator primo comum e n√£o comum, tomamos o maior expoente que aparece na decomposi√ß√£o de qualquer um dos n√∫meros.
   - No exemplo acima, os fatores primos s√£o $2$, $3$, e $5$. Os maiores expoentes s√£o $2^2$, $3^1$, e $5^1$.

3. **Multiplica√ß√£o dos Fatores:**
   - O MMC √© o produto dos fatores escolhidos.
   - $\text{MMC}(12, 15) = 2^2 \times 3^1 \times 5^1 = 4 \times 3 \times 5 = 60$.

##### **Exemplo:**
Vamos calcular o MMC de $8$ e $12$.

1. Decompondo os n√∫meros em fatores primos:
   - $8 = 2^3$
   - $12 = 2^2 \times 3$

2. Tomando os maiores expoentes:
   - $2^3$ (de $8$)
   - $3^1$ (de $12$)

3. Multiplicando os fatores:
   - $\text{MMC}(8, 12) = 2^3 \times 3^1 = 8 \times 3 = 24$.

Portanto, o MMC de $8$ e $12$ √© $24$.

##### **Propriedade Importante:**
Para dois n√∫meros $a$ e $b$, a rela√ß√£o entre o MMC e o MDC (M√°ximo Divisor Comum) √© dada por:

$$
\text{MMC}(a, b) \times \text{MDC}(a, b) = a \times b
$$

Essa propriedade pode ser √∫til para encontrar o MMC se o MDC for conhecido, ou vice-versa.

```python
# Aplica√ß√µes de mmc

# %pip install matplotlib

import matplotlib.pyplot as plt
import math
from functools import reduce

def calcular_mmc(a, b):
    return abs(a * b) // math.gcd(a, b)

def calcular_mmc_de_lista(numeros):
    return reduce(calcular_mmc, numeros)

def gerar_imagem_latex(expressao_latex, nome_arquivo="latex_image.png"):
    plt.figure(figsize=(8, 4))
    plt.text(0.5, 0.5, f"${expressao_latex}$", fontsize=20, ha='center', va='center')
    plt.axis('off')
    plt.savefig(nome_arquivo, bbox_inches='tight', dpi=300)
    plt.show()

# Receber os valores de n n√∫meros
numeros = list(map(int, input("Digite os valores separados por espa√ßo: ").split()))

# Calcular o MMC de todos os n√∫meros inseridos
mmc_total = calcular_mmc_de_lista(numeros)

# Criar uma representa√ß√£o LaTeX da lista de n√∫meros
numeros_latex = ', '.join(map(str, numeros))

# Criar a express√£o LaTeX para o resultado do MMC
expressao_mmc = rf"\mathrm{{MMC}}({numeros_latex}) = {mmc_total}"

# Gerar a imagem com a f√≥rmula do MMC e exibir no notebook
gerar_imagem_latex(expressao_mmc, "mmc_formula.png")
```

#### 2. **M√°ximo Divisor Comum (MDC)**

O M√°ximo Divisor Comum de dois ou mais n√∫meros √© o maior n√∫mero inteiro positivo que divide todos esses n√∫meros sem deixar resto. O MDC √© √∫til em problemas que envolvem simplifica√ß√£o de fra√ß√µes, c√°lculo de propor√ß√µes e alinhamento de divis√µes.

##### **Defini√ß√£o Formal:**
Sejam $a$ e $b$ dois inteiros positivos. O M√°ximo Divisor Comum de $a$ e $b$, denotado por $\text{MDC}(a, b)$, √© o maior n√∫mero inteiro positivo $d$ tal que:

$$
d \mid a \quad \text{e} \quad d \mid b
$$

onde $\mid$ denota "divide".

##### **Como Encontrar o MDC:**
1. **Decomposi√ß√£o em Fatores Primos:**
   - Decompomos cada n√∫mero em seus fatores primos.
   - Exemplo: $18 = 2 \times 3^2$ e $24 = 2^3 \times 3$.

2. **Escolha dos Menores Expoentes:**
   - Para cada fator primo comum, tomamos o menor expoente que aparece na decomposi√ß√£o de ambos os n√∫meros.
   - No exemplo acima, os fatores primos comuns s√£o $2$ e $3$. Os menores expoentes s√£o $2^1$ e $3^1$.

3. **Multiplica√ß√£o dos Fatores:**
   - O MDC √© o produto dos fatores escolhidos.
   - $\text{MDC}(18, 24) = 2^1 \times 3^1 = 2 \times 3 = 6$.

##### **Exemplo:**
Vamos calcular o MDC de $28$ e $35$.

1. Decompondo os n√∫meros em fatores primos:
   - $28 = 2^2 \times 7$
   - $35 = 5 \times 7$

2. Tomando os menores expoentes dos fatores comuns:
   - O fator primo comum √© $7^1$.

3. Multiplicando os fatores:
   - $\text{MDC}(28, 35) = 7^1 = 7$.

Portanto, o MDC de $28$ e $35$ √© $7$.

##### **Algoritmo de Euclides para o MDC:**
O Algoritmo de Euclides √© um m√©todo eficiente para encontrar o MDC de dois n√∫meros inteiros $a$ e $b$. O procedimento √© o seguinte:

1. Divida $a$ por $b$ e obtenha o resto $r$.
2. Substitua $a$ por $b$ e $b$ por $r$.
3. Repita o processo at√© que o resto seja $0$.
4. O MDC √© o √∫ltimo valor de $b$ antes que o resto se torne $0$.

**Exemplo:**
Calcule o MDC de $48$ e $18$ usando o Algoritmo de Euclides.

1. $48 \div 18 = 2$ com resto $12$.
2. Substitu√≠mos $48$ por $18$ e $18$ por $12$.
3. $18 \div 12 = 1$ com resto $6$.
4. Substitu√≠mos $18$ por $12$ e $12$ por $6$.
5. $12 \div 6 = 2$ com resto $0$.
6. O MDC √© $6$.

---

```python
# Aplica√ß√µes de MDC

# %pip install matplotlib
import matplotlib.pyplot as plt
import math
from functools import reduce

def calcular_mdc(a, b):
    return math.gcd(a, b)

def calcular_mdc_de_lista(numeros):
    return reduce(calcular_mdc, numeros)

def gerar_imagem_latex(expressao_latex, nome_arquivo="latex_image.png"):
    plt.figure(figsize=(8, 4))
    plt.text(0.5, 0.5, f"${expressao_latex}$", fontsize=20, ha='center', va='center')
    plt.axis('off')
    plt.savefig(nome_arquivo, bbox_inches='tight', dpi=300)
    plt.show()

# Receber os valores de n n√∫meros
numeros = list(map(int, input("Digite os valores separados por espa√ßo: ").split()))

# Calcular o MDC de todos os n√∫meros inseridos
mdc_total = calcular_mdc_de_lista(numeros)

# Criar uma representa√ß√£o LaTeX da lista de n√∫meros
numeros_latex = ', '.join(map(str, numeros))

# Criar a express√£o LaTeX para o resultado do MDC
expressao_mdc = rf"\mathrm{{MDC}}({numeros_latex}) = {mdc_total}"

# Gerar a imagem com a f√≥rmula do MDC e exibir no notebook
gerar_imagem_latex(expressao_mdc, "mdc_formula.png")
```

### **Rela√ß√£o entre MMC e MDC**

Existe uma rela√ß√£o fundamental entre o MMC e o MDC de dois n√∫meros $a$ e $b$:

$$
\text{MMC}(a, b) \times \text{MDC}(a, b) = a \times b
$$

Essa rela√ß√£o pode ser usada para encontrar o MMC se o MDC for conhecido, ou vice-versa.

**Exemplo:**
Para $a = 12$ e $b = 18$:

1. $\text{MDC}(12, 18) = 6$
2. $12 \times 18 = 216$
3. $\text{MMC}(12, 18) = \frac{216}{6} = 36$

Portanto, $\text{MMC}(12, 18) = 36$ e $\text{MDC}(12, 18) = 6$.

---

```python
# Aplicando rela√ß√£o fundamental

# %pip install matplotlib

import matplotlib.pyplot as plt
import math
from functools import reduce

def calcular_mdc(a, b):
    return math.gcd(a, b)

def calcular_mmc(a, b):
    return abs(a * b) // math.gcd(a, b)

def calcular_mdc_de_lista(numeros):
    return reduce(calcular_mdc, numeros)

def calcular_mmc_de_lista(numeros):
    return reduce(calcular_mmc, numeros)

def gerar_imagem_latex(expressao_latex, nome_arquivo="latex_image.png"):
    plt.figure(figsize=(8, 4))
    plt.text(0.5, 0.5, f"${expressao_latex}$", fontsize=20, ha='center', va='center')
    plt.axis('off')
    plt.savefig(nome_arquivo, bbox_inches='tight', dpi=300)
    plt.show()

# Receber os valores de n n√∫meros
numeros = list(map(int, input("Digite os valores separados por espa√ßo: ").split()))

# Calcular o MDC e o MMC de todos os n√∫meros inseridos
mdc_total = calcular_mdc_de_lista(numeros)
mmc_total = calcular_mmc_de_lista(numeros)

# Criar uma representa√ß√£o LaTeX da lista de n√∫meros
numeros_latex = ', '.join(map(str, numeros))

# Verificar a rela√ß√£o MMC * MDC = Produto dos n√∫meros
produto_total = reduce(lambda x, y: x * y, numeros)
relacao_verificada = mmc_total * mdc_total == produto_total

# Criar a express√£o LaTeX para o resultado do MMC, MDC e a rela√ß√£o
expressao_relacao = rf"\mathrm{{MMC}}({numeros_latex}) \times \mathrm{{MDC}}({numeros_latex}) = {mmc_total} \times {mdc_total} = {produto_total}"

if relacao_verificada:
    expressao_relacao += r" \quad \text{(Rela√ß√£o verificada!)}"
else:
    expressao_relacao += r" \quad \text{(Rela√ß√£o n√£o verificada!)}"

# Gerar a imagem com a f√≥rmula da rela√ß√£o e exibir no notebook
gerar_imagem_latex(expressao_relacao, "relacao_mmc_mdc.png")
```
